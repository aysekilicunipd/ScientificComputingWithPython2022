1\. **Reductions**

Find the total mean, and the mean for each row and column of the following matrix:

```python

```

````python
import numpy as np

m = np.arange(12).reshape((3,4))
print(m, '\n')

total_mean = np.mean(m)
column_mean = np.mean(m, axis = 0)
row_mean    = np.mean(m, axis = 1)

print("Total mean: ", total_mean)
print("Mean of the columns: ", column_mean)
print("Mean of the rows: ", row_mean)
````

2\. **Outer product**

Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
import numpy as np

u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])

#1
print("1. Using the function outer in numpy")
matrix1 = np.outer(u, v)
print(matrix1, "\n")

#2
print("2. Using a nested for loop or a list comprehension")
matrix2 = [i * j for i in u for j in v]
matrix2 = np.array(matrix2)
result = matrix2.reshape(4, 4)
print(result,"\n")

#3
print("3. Using numpy broadcasting operations")
matrix3 = u * v.reshape(4, 1)
print(matrix3.transpose())
````

5\. **Matrices**

Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
import numpy as np

x = range(1, 11) 
y = range(1, 11)

mul_table = [[i * j for j in x] for i in y]
mul_table = np.array(mul_table) 
print(mul_table) 

#Find the trace of the matrix
print("\nTrace:", np.trace(mul_table))

#Extract the anti-diagonal matrix
def anti_diagonal(arr):
    return np.fliplr(arr).diagonal()
print("Anti-diagonal:", anti_diagonal(mul_table))

#Extract the diagonal offset by 1 upwards
print("Diagonal offset by 1:",np.diagonal(mul_table, 1))
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
import matplotlib.pyplot as pyplt
import numpy as np

miles = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
print("Miles: ", miles)

dis = np.abs(miles-miles[:, np.newaxis])
print("Distances in mile:", "\n", dis)

miles_to_km = dis * 1.609344
print("Distances in km:", "\n", miles_to_km.round())
````
